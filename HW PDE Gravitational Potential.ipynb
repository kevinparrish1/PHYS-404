```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy;
omega=1.4

Nmax = 100; Niter = 3000; V = zeros((Nmax, Nmax), float)   


rho=1. 
G=1. 
V[50,50]=4*numpy.pi*G*rho  


for iter in range(Niter):                                  
    if iter%50 == 0: print(iter)
    maxerr=0.
    for i in range(1, Nmax-1): 
        for j in range(1,Nmax-1):
            if i==j==50: continue 
            du = 0.25*(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])-V[i,j]
            if(maxerr<abs(du)):
                maxerr=abs(du)
            V[i,j] =V[i,j] + omega*du
    print(maxerr)
    if(maxerr<1.e-3):
        break

         
x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

def functz(V):                                          
    z = V[X,Y]                        
    return z

Z = functz(V)                          
fig = p.figure()                                      
ax = Axes3D(fig)                                      
ax.plot_wireframe(X, Y, Z, color = 'r')               
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()          
```
