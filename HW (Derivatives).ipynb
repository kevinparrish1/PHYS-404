```python
def derifw(y,x,dx): #forward difference method for a derivative
    return (y(x+dx)-y(x))/dx

import numpy as np
import matplotlib.pyplot as plt
x=(np.pi/4) #set the x value to evaluate cos'(x) at.
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16., -0.5, 0.1): #create a loop to store 10^-16,10^-15,...,10^-1 in an array
    y=derifw(np.cos,x,dx) 
    dxarr.append(dx) #stores the dx value used
    yarr.append(y) #stores the numerically evaluated y value using the specific dx
    
plt.plot(dxarr, abs(yarr+np.sin(x)), 'b-', lw=2) #plots the error (defined as true value-numerically calculated value) vs the corresponding dx value    
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')

plt.plot(x,10^-5)

plt.show()
#the error as Error~dx^n shows that n is 1. It is 1st order. The optimal value for dx is around 10^-7
```

```python
def dericd(y,x,dx): #central difference method for a derivative
    return (y(x+dx/2)-y(x-dx/2))/dx

import numpy as np
import matplotlib.pyplot as plt
x=(np.pi/4) #set the x value to evaluate cos'(x) at.
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16., -0.5, 0.1): #create a loop to store 10^-16,10^-15,...,10^-1 in an array
    y=dericd(np.cos,x,dx) 
    dxarr.append(dx) #stores the dx value used
    yarr.append(y) #stores the numerically evaluated y value using the specific dx
    
plt.plot(dxarr, abs(yarr+np.sin(x)), 'b-', lw=2) #plots the error (defined as true value-numerically calculated value) vs the corresponding dx value
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')

plt.show()


#This scheme is much better because not only is the optimal value for dx larger (thus making it more
#manageable), you are also able to achieve much greater accuracy (the error value is smaller here)
```
